```python
# --- STEP 4: SERVE DETECTION ---
# A "Serve" is defined by entering the Strike Zone (>150 degrees)
strike_threshold = 150
serves = []
in_serve = False

for i in range(1, len(df)):
    if df.iloc[i]['Abduction_Angle'] >= strike_threshold and not in_serve:
        in_serve = True
        start_t = df.iloc[i]['Seconds']
    elif df.iloc[i]['Abduction_Angle'] < 120 and in_serve:
        in_serve = False
        end_t = df.iloc[i]['Seconds']

        # Analyze this specific serve
        serve_segment = df[(df['Seconds'] >= start_t) & (df['Seconds'] <= end_t)]
        max_angle = serve_segment['Abduction_Angle'].max()

        serves.append({'Time': start_t, 'Peak': max_angle})

# --- STEP 5: THE REPORT ---
print(f"🎾 TENNIS SERVE REPORT")
print(f"Total Serves Detected: {len(serves)}")
print("-" * 30)
for i, s in enumerate(serves):
    print(f"Serve {i+1}: Peak Reach {s['Peak']:.1f}° at {s['Time']:.2f}s")
```

```python
# --- STEP 6: VELOCITY & SMOOTHNESS ---
print(f"🚀 SERVE VELOCITY ANALYSIS")
print("-" * 30)

for i, s in enumerate(serves):
    # Look at the 0.5 seconds before the peak to find the 'strike speed'
    strike_window = df[(df['Seconds'] >= s['Time'] - 0.5) & (df['Seconds'] <= s['Time'])]

    if len(strike_window) > 1:
        # Calculate degrees per second
        deg_change = strike_window['Abduction_Angle'].iloc[-1] - strike_window['Abduction_Angle'].iloc[0]
        time_change = strike_window['Seconds'].iloc[-1] - strike_window['Seconds'].iloc[0]
        velocity = deg_change / time_change

        # Rate the intensity
        intensity = "🔥 HIGH POWER" if velocity > 150 else "⚡ MODERATE"
        print(f"Serve {i+1}: {velocity:.1f} deg/sec | {intensity}")
    else:
        print(f"Serve {i+1}: Data gap - unable to calculate.")
```

```python
# --- STEP 7: SMOOTHNESS & JITTER ANALYSIS ---
import numpy as np

print(f"📊 SHOULDER STABILITY REPORT")
print("-" * 30)

for i, s in enumerate(serves):
    # Analyze the 1 second window during and after the strike
    window = df[(df['Seconds'] >= s['Time']) & (df['Seconds'] <= s['Time'] + 1.0)]

    if len(window) > 1:
        # Calculate 'Jitter' (Mean Absolute Successive Difference)
        # This measures how much the angle "wobbles" line-to-line
        diffs = np.abs(np.diff(window['Abduction_Angle']))
        jitter = np.mean(diffs)

        # Rating the stability
        if jitter < 1.5:
            rating = "💎 STABLE"
        elif jitter < 3.0:
            rating = "⚖️ NEUTRAL"
        else:
            rating = "⚠️ SHAKY (Watch Form)"

        print(f"Serve {i+1}: {jitter:.2f} Jitter Score | {rating}")
```

```python
# --- STEP 8: RHYTHM & TIMING ---
print(f"⏱️ SERVE RHYTHM ANALYSIS")
print("-" * 30)

rhythms = []

for i, s in enumerate(serves):
    # Find the last time the arm was in the 'Toss Zone' (110 deg) before this strike
    toss_data = df[(df['Seconds'] < s['Time']) & (df['Abduction_Angle'] >= 80) & (df['Abduction_Angle'] <= 110)]

    if not toss_data.empty:
        toss_time = toss_data['Seconds'].iloc[-1]
        rhythm_duration = s['Time'] - toss_time
        rhythms.append(rhythm_duration)
        print(f"Serve {i+1}: Toss-to-Strike = {rhythm_duration:.2f}s")
    else:
        print(f"Serve {i+1}: Toss phase not detected.")

if rhythms:
    consistency = np.std(rhythms)
    print("-" * 30)
    print(f"Rhythm Variance: {consistency:.3f}s")
    if consistency < 0.1:
        print("Result: 🟢 PRO-LEVEL CONSISTENCY")
    else:
        print("Result: 🟡 TIMING FLUCTUATION")
```
